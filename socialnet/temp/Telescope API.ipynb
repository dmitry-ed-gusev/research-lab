```python
assignmentIds = []
assignerTypes = []
badgeIds = []
fromNames = []
fromIds = []
dates = []
comments = []
upsaIds = []
```

```python
for rec in heroes_loaded_list:
    for badge in rec.get('bean').get('badge'):
        assignmentIds.append(badge.get('assignmentId'))
        assignerTypes.append(badge.get('assignerType'))
        badgeIds.append(badge.get('id'))
        fromNames.append(badge.get('fromName'))
        fromIds.append(badge.get('fromId'))
        dates.append(badge.get('date'))
        comments.append(badge.get('comment'))
        upsaIds.append(rec.get('bean').get('upsaId'))
```

```python
heroes_df = pd.DataFrame({'assignmentId': assignmentIds,
                         'assignerType': assignerTypes,
                         'badgeId': badgeIds,
                         'fromName': fromNames,
                         'fromId': fromIds,
                         'date': dates,
                         'comment': comments,
                         'upsaId': upsaIds})
```
